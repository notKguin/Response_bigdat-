##  Задание 3:

Выведите сколько у каждой кафедры закреплено электронных курсов на портале. 

У какой кафедры больше всего курсов на портале?

## Задание 5:

Выполните задание 3, но теперь нужно вывести не код кафедры, а ее название и количество курсов, которые за ней закреплены. 

У какой кафедры больше всего курсов?

````python
%%sql postgresql:///test

SELECT d.name, COUNT(DISTINCT l.courseid) AS course_count
FROM logs l
JOIN departments d ON l.depart::smallint = d.id
GROUP BY d.name;
````

## Задание 6:

Ответьте на вопрос: существуют ли курсы, за которыми закреплено несколько кафедр? Если такие курсы есть, то выведите их количество.
Также выведите названия кафедр, которые совместно преподают один и тот же курс.




````python
%%sql postgresql:///test

SELECT COUNT(DISTINCT courseid)
FROM (
    SELECT courseid
    FROM logs
    GROUP BY courseid
    HAVING COUNT(DISTINCT Depart) > 1
) AS subquery;
````

````python
%%sql postgresql:///test

SELECT l.courseid, STRING_AGG(DISTINCT d.name, ', ') AS department_names
FROM logs l
JOIN departments d ON l.Depart::smallint = d.id
WHERE l.courseid IN (
    SELECT courseid
    FROM logs
    GROUP BY courseid
    HAVING COUNT(DISTINCT Depart) > 1
)
GROUP BY l.courseid;
````

## Задание 7:

Выведите количество студентов, которые получили 2, 3, 4, 5.

Пример вывода:

| namer_level |	count |
|-----|------|
|2 |	4 |
|3 |	3435 |
|4 | 	4676765|
|5 | 232 |


````python
%%sql postgresql:///test

SELECT NameR_Level AS grade, COUNT(DISTINCT userid) AS count
FROM logs
WHERE NameR_Level IN ('2', '3', '4', '5')
GROUP BY NameR_Level;
````

## Задание 8:

Выведите студента, который больше всех работает на портале (у него максимальное количество логов за вест период обучения).

````python
%%sql postgresql:///test

SELECT userid, SUM(s_all) AS total_logs
FROM logs
GROUP BY userid
ORDER BY total_logs DESC
LIMIT 1;
````

## Задание 9:

Выведите по каждой недели среднее количество всех событий на портале.

````python
%%sql postgresql:///test

SELECT num_week, AVG(s_all) AS average_events
FROM logs
GROUP BY num_week
ORDER BY num_week;
````

## Задание 10: 

Выведите название кафедры, у которой больше всего отличников.

Отдельно выведите название кафедры, у которой больше всего двоечников. 

````python
%%sql postgresql:///test

SELECT d.name, COUNT(DISTINCT l.userid) AS count_top_students
FROM logs l
JOIN departments d ON l.Depart::smallint = d.id
WHERE l.NameR_Level = '5'
GROUP BY d.name
ORDER BY count_top_students DESC
LIMIT 1;
````

````python
%%sql postgresql:///test

SELECT d.name, COUNT(DISTINCT l.userid) AS count_low_students
FROM logs l
JOIN departments d ON l.Depart::smallint = d.id
WHERE l.NameR_Level = '2'
GROUP BY d.name
ORDER BY count_low_students DESC
LIMIT 1;
````
